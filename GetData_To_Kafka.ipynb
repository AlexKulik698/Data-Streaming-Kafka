```python
import yliveticker
import pandas as pd
import requests
import string
import json
import time 
from datetime import date, datetime
```

```python
## VERY important to pass the bearer for authentication
my_headers = {"Content-Type" : "application/json", "Bearer" : "ea5afb228d2a4662bf80eb824ac43629"}
```

```python
# function to consume the data from ticker and send the data to the api
def on_new_msg(ws, msg):
    # Extract the relevant fields from the message
    data = {        
        'tickerID': str(msg['id']),
        'exchange': str(msg['exchange']),
        'quoteType': str(msg['quoteType']),
        'price': str(msg['price']),
        'yt_timestamp': str(msg['timestamp']),
        'marketHours': str(msg['marketHours']),
        'changePercent': str(msg['changePercent']),
        'dayVolume': str(msg['dayVolume']),
        'change': str(msg['change']),
        'priceHint': str(msg['priceHint'])
    }
    
    # Append the data to the DataFrame
    df = pd.DataFrame(columns=['tickerID', 'exchange', 'quoteType', 'price', 'yt_timestamp', 'marketHours', 'changePercent', 'dayVolume', 'change', 'priceHint'])
    df = pd.concat([df, pd.DataFrame(data, index=[0])], ignore_index=True)

    body = df[["yt_timestamp","tickerID","exchange","quoteType","price","marketHours","changePercent","dayVolume","change","priceHint"]].astype("str").to_json(orient='records')

    # just make sure data is working
    print(body)
    
    response = requests.post("http://ds04-producer.dnsdojo.org/api/v1/Producer", headers=my_headers, data=body, timeout=5)    
    print(f'{datetime.now()} : {response.status_code}: {response.reason}')
    
    # reset the data frame. Yes i know its worse way to do this, but this is just a POC :) 
    del df
    
```

```python
# Create a YLiveTicker object with the on_new_msg callback function
# GM = General Motors
yliveticker.YLiveTicker(on_ticker=on_new_msg, ticker_names=['GM'])
```
